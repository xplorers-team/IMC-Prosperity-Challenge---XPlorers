```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_0 = pd.read_csv(filepath_or_buffer='../data/prices_round_3_day_0.csv', sep=';')
data_1 = pd.read_csv(filepath_or_buffer='../data/prices_round_3_day_1.csv', sep=';')
data_2 = pd.read_csv(filepath_or_buffer='../data/prices_round_3_day_2.csv', sep=';')
data = pd.concat([data_0, data_1, data_2], axis=0).reset_index(drop=True)
```

```python
data = pd.read_csv(
    '../data/prices_round_3_day_1.csv',
    sep = ';'
)
```

```python
def plot_price_trends(product_name):
    product_data = data[data['product'] == product_name]
    
    plt.figure(figsize=(14, 7))
    plt.plot(product_data['timestamp'], product_data['bid_price_1'], label='Top Bid Price')
    plt.plot(product_data['timestamp'], product_data['ask_price_1'], label='Top Ask Price')
    plt.plot(product_data['timestamp'], product_data['mid_price'], label='Mid Price', color='red', linewidth=2)
    plt.title(f'Price Trends for {product_name}')
    plt.xlabel('Timestamp (milliseconds)')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot price trends for "CHOCOLATE"
plot_price_trends('CHOCOLATE')

# Function to plot volume trends for a specific product using timestamps
def plot_volume_trends(product_name):
    product_data = data[data['product'] == product_name]
    
    plt.figure(figsize=(14, 7))
    plt.bar(product_data['timestamp'], product_data['bid_volume_1'], width=100, label='Top Bid Volume', color='blue')
    plt.bar(product_data['timestamp'], product_data['ask_volume_1'], width=100, label='Top Ask Volume', color='orange')
    plt.title(f'Volume Trends for {product_name}')
    plt.xlabel('Timestamp (milliseconds)')
    plt.ylabel('Volume')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot volume trends for "CHOCOLATE"
plot_volume_trends('CHOCOLATE')

# Comparing Mid Prices of Different Products using timestamps
products_to_compare = ['CHOCOLATE', 'STRAWBERRIES', 'ROSES', 'GIFT_BASKET']
comparison_data = data[data['product'].isin(products_to_compare)]

fig, ax = plt.subplots(figsize=(14, 7))
for product in products_to_compare:
    product_data = comparison_data[comparison_data['product'] == product]
    ax.plot(product_data['timestamp'], product_data['mid_price'], label=f'Mid Price of {product}')

ax.set_title('Comparison of Mid Prices Across Products')
ax.set_xlabel('Timestamp (milliseconds)')
ax.set_ylabel('Mid Price')
ax.legend()
ax.grid(True)
plt.show()
```

```python
window = 400

avg_rolling_spread = spread.rolling(window)\
    .mean()

mov_spread5 = spread.rolling(5).maean()

rolling_std_spread = spread.rolling(window)\
    .std()
```

```python
zscore_spread = (mov_spread5 - avg_rolling_spread)/rolling_std_spread
```

```python
zscore_spread.plot()
```

```python
buy = zscore_spread < -2
sell = zscore_spread > 2
```

```python
spread.plot(label = 'spread')
spread[buy].plot(
    marker ='^', 
    color ='green', 
    linestyle= 'None',
    label = 'buy'
)
spread[sell].plot(
    marker= '^', 
    color = 'red', 
    linestyle= 'None',
    label = 'sell'
)

plt.legend()
plt.xlim([0, 1e6])
```

```python
position_size = (spread[buy].shape[0] - spread[sell].shape[0])
position_size
```

```python
pnl = (spread[sell].sum() - spread[buy].sum() - position_size*spread.iloc[-1])
```

```python
position = pd.Series(0, index = gift_basket.index)
```

```python
for idx, value in spread[sell].items():
    position.loc[idx] = -1

for idx, value in spread[buy].items():
    position.loc[idx] = 1

position = position.cumsum()
```

```python
cash = pd.concat([spread[sell], -spread[buy]])\
    .sort_index()\
    .cumsum()\
    .reindex(gift_basket.index)\
    .fillna(method="ffill")
```

```python
value_on_position = (position * spread)\
    .reindex(gift_basket.index)\
    .fillna(method="ffill")
```

```python
pnl = value_on_position + cash
```

```python
pnl.plot(
    label = "pnl"
)
plt.grid()
plt.legend()


plt.figure()
position.plot(
    label = "number of active positions"
)
plt.grid()
plt.legend()
```
