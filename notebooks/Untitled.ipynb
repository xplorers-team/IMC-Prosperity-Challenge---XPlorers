```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data0 = pd.read_csv(filepath_or_buffer='../data/prices_round_3_day_0.csv', sep=';')
data1 = pd.read_csv(filepath_or_buffer='../data/prices_round_3_day_1.csv', sep=';')
data2 = pd.read_csv(filepath_or_buffer='../data/prices_round_3_day_2.csv', sep=';')
```

```python
chocolate = data0.query("product == 'CHOCOLATE'")\
    .set_index("timestamp")
strawberries = data0.query("product == 'STRAWBERRIES'")\
    .set_index("timestamp")
roses = data2.query("product == 'ROSES'")\
    .set_index("timestamp")
gift_basket = data1.query("product == 'GIFT_BASKET'")\
    .set_index("timestamp")
basket_separated = (6*strawberries + 4*chocolate + roses)\
    .mid_price
```
