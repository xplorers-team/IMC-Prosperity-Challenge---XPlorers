```python
import numpy as np

import pandas as pd
import scipy.stats as stats
from scipy.stats import norm
from scipy.stats import skew

import matplotlib.pyplot as plt
```

```python

def monte_carlo_option_price(S, K, T, r, sigma, num_simulations=10000):
    # Generate random draws from a standard normal distribution
    Z = np.random.normal(size=num_simulations)
    
    # Calculate the asset price at expiration for each simulation
    ST = S * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)
    
    # Calculate the payoff for each simulation
    payoffs = np.maximum(ST - K, 0)
    
    # Calculate the average payoff
    average_payoff = np.mean(payoffs)
    
    # Discount the average payoff to present value
    option_price = np.exp(-r * T) * average_payoff
    
    return option_price
```

```python
#BLACK AND SCHOLES 
def black_scholes_call(S, K, T, r, vol):
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = d1 - vol * np.sqrt(T)
    call_price = (S * stats.norm.cdf(d1) - K * np.exp(-r * T) * stats.norm.cdf(d2))
    return call_price

def calculate_rolling_volatility(prices, window):
        # Assuming prices is a Series of mid prices
        log_returns = np.log(prices / prices.shift(1))
        rolling_volatility = log_returns.rolling(window=window).std() * np.sqrt(252)  # annualized volatility
        return rolling_volatility
```

```python
# Parameters
expiration_day = 246  # This is the final day of the option period
total_period_days = 365  # Total number of days in the option period
average_spread = 0  # This is the average spread you've determined


# Add a 'day' column to your DataFrame by scaling down the 'timestamp'
full_data['day'] = full_data['timestamp'] / 1000000

# Now calculate the time to expiration 'T' for each entry in the DataFrame
full_data['T'] = (expiration_day - full_data['day']) / total_period_days
K = 10000  # strike price
coconut_data['T'] = (expiration_day - coconut_data['day']) / total_period_days
r = 0      # risk-free rate
vol = 0.19314 # volatility of the underlying asset

# Filter for COCONUT and calculate Black-Scholes prices
coconut_data = full_data[full_data['product'] == 'COCONUT'].copy()
coconut_data['mid_price'] = pd.to_numeric(coconut_data['mid_price'], errors='coerce')
coconut_data['log_returns'] = np.log(coconut_data['mid_price'] / coconut_data['mid_price'].shift(1))

print(coconut_data['mid_price'].std())



# Calculate the skewness of the log returns
log_returns_skewness = skew(coconut_data['log_returns'].dropna())

print(f"Skewness of COCONUT's log returns: {log_returns_skewness}")

# Adjust the volatility based on the skewness
# This is a heuristic method and not a standard approach
#vol_adjusted_for_skewness = vol * (1 + 0.05* log_returns_skewness)  # `beta` is a sensitivity factor to skewness


coconut_data['bs_price'] = coconut_data.apply(lambda row: black_scholes_call(row['mid_price'], K, row['T'], r, vol) if pd.notnull(row['mid_price']) else None, axis=1)

# Subtract the average spread from the Black-Scholes price
coconut_data['bs_price_adjusted'] = coconut_data['bs_price'] - average_spread
# Filter for COCONUT_COUPON
coconut_coupon_data = full_data[full_data['product'] == 'COCONUT_COUPON'].copy()
coconut_coupon_data['mid_price'] = pd.to_numeric(coconut_coupon_data['mid_price'], errors='coerce')

# Ensure that the timestamps match between COCONUT and COCONUT_COUPON
# This might involve some data preprocessing to align timestamps if necessary

# Plot the Black-Scholes prices and actual COCONUT_COUPON prices against timestamps
plt.figure(figsize=(14, 7))
plt.plot(coconut_data['timestamp'], coconut_data['bs_price_adjusted'], label='Black-Scholes Price', linewidth=1.6)  # Set a thinner line
plt.plot(coconut_coupon_data['timestamp'], coconut_coupon_data['mid_price'], label='COCONUT_COUPON Mid Price', linewidth=1.6)  # Set a thinner line

plt.title('COCONUT_COUPON Prices Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
