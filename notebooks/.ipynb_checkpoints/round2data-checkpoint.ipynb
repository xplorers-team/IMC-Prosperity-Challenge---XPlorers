```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_0 = pd.read_csv(filepath_or_buffer='../data/prices_round_2_day_-1.csv', sep=';')
data_1 = pd.read_csv(filepath_or_buffer='../data/prices_round_2_day_0.csv', sep=';')
data_2 = pd.read_csv(filepath_or_buffer='../data/prices_round_2_day_1.csv', sep=';')
data = pd.concat([data_0, data_1, data_2], axis=0).reset_index(drop=True)
```

```python
data_0.head()
```

```python
data_0.plot(x='timestamp', y='ORCHIDS')
plt.show()
```

```python
data_0.set_index('timestamp', inplace=True)

# Create a figure and a single subplot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the Orchids price data
ax1.plot(data_0.index, data_0['ORCHIDS'], label='Orchids Price', color='blue')
ax1.set_xlabel('Timestamp')
ax1.set_ylabel('Orchids Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create another y-axis that shares the same x-axis
ax2 = ax1.twinx()  
ax2.plot(data_0.index, data_0['HUMIDITY'], label='Humidity', color='green', linestyle='--')
ax2.set_ylabel('Humidity', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Title and legend
plt.title('Orchids Price and Humidity Over Time')
fig.tight_layout()  # Adjust layout to make room for the y-labels
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the Orchids price data
ax1.plot(data_0['ORCHIDS'], label='Orchids Price', color='blue')
ax1.set_xlabel('Timestamp')
ax1.set_ylabel('Orchids Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create another y-axis that shares the same x-axis
ax2 = ax1.twinx()
ax2.plot(data_0['SUNLIGHT'], label='Sunlight', color='orange', linestyle='--')
ax2.set_ylabel('Sunlight', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Title and legend
plt.title('Orchids Price and Sunlight Over Time')
fig.tight_layout()  # Adjust layout to make room for the y-labels
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
# Create a figure and a single subplot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot Sunlight on the primary y-axis (left)
ax1.plot(data_0.index, data_0['SUNLIGHT'], label='Sunlight', color='orange')
ax1.set_xlabel('Timestamp')
ax1.set_ylabel('Sunlight', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')

# Create another y-axis for Humidity using the same x-axis
ax2 = ax1.twinx()
ax2.plot(data_0.index, data_0['HUMIDITY'], label='Humidity', color='green')
ax2.set_ylabel('Humidity', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Adding a title to the plot
plt.title('Sunlight and Humidity Over Time')

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the Orchids price on the primary y-axis (left)
ax1.plot(data_0.index, data_0['ORCHIDS'], label='Orchids Price', color='blue')
ax1.set_xlabel('Timestamp')
ax1.set_ylabel('Orchids Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create another y-axis for Import Tariffs using the same x-axis
ax2 = ax1.twinx()
ax2.plot(data_0.index, data_0['IMPORT_TARIFF'], label='Import Tariff', color='red')
ax2.set_ylabel('Import Tariff', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Adding a title to the plot
plt.title('Orchids Price and Import Tariff Over Time')

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
average_env = (data_0['SUNLIGHT'] + data_0['HUMIDITY']) / 2

# Create a figure and a single subplot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the 'average' environment curve
ax1.plot(data_0.index, average_env, label='Average Sunlight & Humidity', color='purple')
ax1.set_xlabel('Timestamp')
ax1.set_ylabel('Average Sunlight & Humidity', color='purple')
ax1.tick_params(axis='y', labelcolor='purple')
ax1.legend(loc='lower left')

# Create another y-axis for the Orchid price using the same x-axis
ax2 = ax1.twinx()
ax2.plot(data_0.index, data_0['ORCHIDS'], label='Orchids Price', color='blue')
ax2.set_ylabel('Orchids Price', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.legend(loc='upper left')

# Add a title to the plot
plt.title('Average Environmental Factors and Orchids Price Over Time')

# Show the plot
plt.show()
```

```python
# Assuming data_0 is your DataFrame and it has 'HUMIDITY' and 'ORCHIDS' columns
# First, filter the data where humidity is below 60%
low_humidity_data = data_0[data_0['HUMIDITY'] > 80]

# Now create a plot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the filtered Orchid prices
ax1.plot(low_humidity_data.index, low_humidity_data['ORCHIDS'], label='Orchids Price', color='blue')
ax1.set_xlabel('Timestamp')
ax1.set_ylabel('Orchids Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')

# Create another y-axis for the humidity levels using the same x-axis
ax2 = ax1.twinx()
ax2.plot(low_humidity_data.index, low_humidity_data['HUMIDITY'], label='Humidity', color='green', linestyle='--')
ax2.set_ylabel('Humidity (%)', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.legend(loc='upper right')

# Add a title to the plot
plt.title('Orchids Price and Humidity Levels When Humidity is Below 60%')

# Show the plot
plt.show()
```
